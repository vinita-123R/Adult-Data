#### There are some ? that canbe seen in data, checkit and get rid of it

```python
# getting theonlythevalue of work class is ? and assigning it to df1
df1=df[df['workclass']=='?']
df1.head()
```

```python
# getting the data which doesn't have ?
df=df[df['workclass']!='?']
df.head()
```

```python
# select all categorical variables
df_categorical=df.select_dtypes(include=['object'])

# checking weather any other columns containa ?
df_categorical.apply(lambda x:x=='?',axis=0).sum()
```

```python
# dropping the ?
df=df[df['occupation']!='?']
df=df[df['native.country']!='?']
```

###  Apply label encoding to the new clean dataframe

```python
from sklearn import preprocessing
# encode categorical variables using label encoder
# select all categorical variables
df_cateorical=df.select_dtypes(include=['object'])
df_categorical.head()
```

```python
# apply label encoder to df_cateorical
le=preprocessing.LabelEncoder()
df_categorical=df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
# convert target variable income to categorical
df_categorical['income']=df_categorical['income'].astype('category')
df_categorical.info()
```

### splitting the data into train and test

```python
# putting feature variable to x
X=df_categorical.drop('income',axis=1)

# putting response variable to y
y=df_categorical['income']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=99)

X_train.head()
```

### Build a decision tree and print confusion matrix

```python
from sklearn.tree import DecisionTreeClassifier

dt_default=DecisionTreeClassifier(max_depth=5)

dt_default.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
y_pred_default=dt_default.predict(X_test)
print(classification_report(y_test,y_pred_default))
```

```python
print(confusion_matrix(y_test,y_pred_default))
print(accuracy_score(y_test,y_pred_default))
```

### Use grid search to find the best params for Decision tree classifier

```python
# Grid search cv to find optimal max_depth
# from sklearn.model_selection import k-fold

from sklearn.model_selection import GridSearchCV
```

```python
# specify number of folds for k-fold CV
n_folds=5

# parameters to build model on
parameters={'max_depth':range(1,10),
           'min_samples_split':range(1,8),
           'ccp_alpha':[0,0.1,0.3,0.55]}
```

```python
# instantiate the model
dtree=DecisionTreeClassifier(criterion='gini',random_state=100)
```

```python
# fit tree on the training data
tree=GridSearchCV(dtree,parameters,cv=n_folds,scoring='accuracy')
tree.fit(X_train,y_train)
```

```python
# scores of GridSearch CV
scores=tree.cv_results_
pd.DataFrame(scores).head(20)
```

```python
param=tree.best_params_
param
```

```python
tree.best_score_
```
